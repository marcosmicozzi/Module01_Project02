# Grouping

```python
print('Grouping Red Wine')
# create new column
red_wine['quality_label'] = red_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# transform into categorical type
red_wine['quality_label'] = pd.Categorical(red_wine['quality_label'], categories=['low', 'medium', 'high'])

print(red_wine['quality_label'].value_counts())
```

```python
print('Grouping White Wine')
# create new column
white_wine['quality_label'] = white_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# transform into categorical type
white_wine['quality_label'] = pd.Categorical(white_wine['quality_label'], categories=['low', 'medium', 'high'])

print(white_wine['quality_label'].value_counts())
```

# Merging Both Datasets

```python
red_wine['wine_type'] = 'red wine'
white_wine['wine_type'] = 'white wine'

```

```python
both_lists = [red_wine, white_wine]

all_wines = pd.concat(both_lists)
print('New Data Frame')
all_wines['wine_type']
```

```python
# Compute the correlation matrix (numerical columns only)
corr = red_wine.corr(numeric_only=True)

# Plot the heatmap
# I had to add fig and ax to get it to the right size. why!!!!!
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(data=corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.1, square = True, vmin=-1, vmax=1, ax=ax)
plt.title("Correlation Heatmap of Red Wine Data")


np.triu(np.ones_like(corr, dtype=bool))

plt.show()
```
