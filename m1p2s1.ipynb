# Grouping

```python
print('Grouping Red Wine')
# create new column
red_wine['quality_label'] = red_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# transform into categorical type
red_wine['quality_label'] = pd.Categorical(red_wine['quality_label'], categories=['low', 'medium', 'high'])

print(red_wine['quality_label'].value_counts())
```

```python
print('Grouping White Wine')
# create new column
white_wine['quality_label'] = white_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# transform into categorical type
white_wine['quality_label'] = pd.Categorical(white_wine['quality_label'], categories=['low', 'medium', 'high'])

print(white_wine['quality_label'].value_counts())
```

# Epic 3: Statistical Significance


```python
from scipy import stats

F, p = stats.f_oneway(
    red_wine[red_wine['quality_label'] == 'low']['alcohol'],
    red_wine[red_wine['quality_label'] == 'medium']['alcohol'],
    red_wine[red_wine['quality_label'] == 'high']['alcohol']
)
print('ANOVA test for mean alcohol levels across wine samples with different quality ratings')
print('F Statistic:', F, '\tp-value:', p)
```

```python
# ANOVA Test Chlorides and Quality level.

from scipy import stats

F, p = stats.f_oneway(
    red_wine[red_wine['quality_label'] == 'low']['chlorides'],
    red_wine[red_wine['quality_label'] == 'medium']['chlorides'],
    red_wine[red_wine['quality_label'] == 'high']['chlorides']
)
print('ANOVA test for mean wine types across wine samples with different quality ratings')
print('F Statistic:', F, '\tp-value:', p)
```

```python
# ANOVA wine type and chlorides RESULT 0.0

from scipy import stats

F, p = stats.f_oneway(
    all_wines[all_wines['wine_type'] == 'red wine']['chlorides'],
    all_wines[all_wines['wine_type'] == 'white wine']['chlorides'],
)
print('ANOVA test for mean wine types across wine samples with different quality ratings')
print('F Statistic:', F, '\tp-value:', p)
```

```python
# ANOVA wine type and residual sugar

from scipy import stats

F, p = stats.f_oneway(
    all_wines[all_wines['wine_type'] == 'red wine']['residual_sugar'],
    all_wines[all_wines['wine_type'] == 'white wine']['residual_sugar'],
)
print('ANOVA test for mean wine types across wine samples with different quality ratings')
print('F Statistic:', F, '\tp-value:', p)
```

```python
# ANOVA wine type and quality

from scipy import stats

F, p = stats.f_oneway(
    all_wines[all_wines['wine_type'] == 'red wine']['quality'],
    all_wines[all_wines['wine_type'] == 'white wine']['quality'],
)
print('ANOVA test for mean wine types across wine samples with different quality ratings')
print('F Statistic:', F, '\tp-value:', p)
```
