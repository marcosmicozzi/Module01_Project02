## Step 1: Import Matplotlib Library

```python
# pyplot stands for python and plot / plt is a shortcut
import matplotlib.pyplot as plt
import numpy as np
```

## Step 2: Initialize the Chart

```python
# fig: creates the overall canvas where everything is drawn
# ax: the area or section within that canvas where the actual graph appears
fig, ax = plt.subplots()
```

## Step 3: Fill the axes (`ax`)
The `plot()` method takes several arguments. In this case, we've provided two lists: one for the x-values and one for the y-values of each data point.

```python
ax.plot([1,2,3,4,5], [2,3,1,4,5])
plt.show()
```

## Step 4: Reveal the plot with `show()`

```python
# all together
x = [1,2,3,4,5]
y = [2,3,1,4,5]

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

### Subplots 
Subplots enable us to plot side by side, or in several rows and columns.

```python
# one row / 2 columns
fig, ax = plt.subplots(nrows=1, ncols=2)
# fig, ax = plt.subplots(1,2) # another way to write it
# define index
ax[0].plot([1, 2, 3, 4, 5], [2, 3, 1, 4, 5], color="blue")
ax[1].scatter([1, 2, 3, 4, 5], [2, 3, 1, 4, 5], color="blue")

plt.show()
```

```python
# Sample data
x = np.linspace(1,11,20) # creates 20 numbers between 1 and 11
y = x**2

# We can also adjust the fig size (h,v)
fig, ax = plt.subplots(figsize=(8, 5), nrows=2, ncols=3)

# For multiple rows, axes is a 2D array → use axes[row, col] to access subplots
ax[0, 0].scatter(x, y, color="red")  # Top-left
ax[1, 2].scatter(x, y, color="purple")  # Bottom-right

# You can also iterate over all axes in a flat loop / .ravel() is flattening the array
for ax in ax.ravel():
    ax.plot(x, y)
    ax.set_title("Wow!")


plt.tight_layout()

```

## Pyplot vs. Axes

There are two ways to create plots in Matplotlib:

- The `Pyplot interface` (functional method) lets you create plots with slightly **less code** (usually 1 or 2 fewer lines), making it ideal for quickly generating simple charts.

- The `Axes interface` (object-oriented method) requires initializing your chart with `fig, axes = plt.subplots()` before plotting. It uses a more object-oriented approach, offering **greater customization and control** over your visualizations.

**The Axes Interface covers 95% of typical use cases - We advice to use Axes Interface!**

```python
# Pyplot
plt.plot([1,2,3,4,5], [2,3,1,4,5])
plt.show()

# Axes with subplots
fig, ax = plt.subplots()
ax.plot([1,2,3,4,5], [2,3,1,4,5])
plt.show()
```

## Special Case `plt.figure()`
Gives even more customizability on the axes placement within the figure, i.e. for a zoomed in view


```python
# # Axes with subplots
# fig, ax = plt.subplots()
# ax.plot([1,2,3,4,5], [2,3,1,4,5])
# plt.show()

# Axes with figure
fig = plt.figure(figsize=(8,6))
ax = fig.add_axes([0.2, 40, 0.5, 0.7]) # tbc
ax.plot(x,y)
plt.show()
```

```python
# Axes with figure
fig = plt.figure(figsize=(8,6))

ax1 = fig.add_axes([0.0, 0.0, 0.8, 0.8])
ax2 = fig.add_axes([-0.3, 0.6, 0.2, 0.2]) # x-axis, y-axis, width, height

ax1.plot(x,y)
ax2.plot(y,x)

ax1.set_title('Larg Plot')
ax2.set_title('Small Plot')

plt.show()
```

## Figure, Axes, and Subplot
Figure, Axes, and Subplots are the **three core components** of any Matplotlib chart. 
- `Figure` is the overall canvas where everything is drawn
- `Axes` are "plot container" where the actual graph happens 

Every plot lives inside an axes, and each figure can contain multiple axes to organize different visualizations.









## Set Titles & Labels
- `set_title()`
- `set_xlabel()`
- `set_ylabel()`


```python
# Axes with subplots
fig, ax = plt.subplots()
ax1 = ax.plot([1,2,3,4,5], [2,3,1,4,5], label = "ax1")
ax2 = ax.plot([2,3,1,4,5], [1,2,3,4,5], label = "ax2")

ax.set_title("Title")
ax.set_xlabel("X-Label")
ax.set_ylabel("Y-Label")
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots(nrows= 1, ncols = 2)

ax[0].plot(x,y)
ax[0].set_title('1st Plot')

ax[1].plot(y,x)
ax[1].set_title('2nd Plot')

fig.suptitle('My Super Title', fontsize=16)  # Large title above all subplots

plt.show()
```

## Save figure as png



```python
fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x, y)
ax[0].set_title('1st Plot')

ax[1].plot(y, x)
ax[1].set_title('2nd Plot')

fig.suptitle('My Super Title', fontsize=16)

# Save with regular and transparent background
fig.savefig('myfig.png', dpi=200)
fig.savefig('myfig_transparentNew.png', dpi=500, transparent=True)

# Show plot
plt.show()
```

## Add Legend

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3, 4, 5], [2, 3, 1, 4, 5], 'r', label="Red Line")
ax.plot([1, 2, 3, 4, 5], [1,5,2, 4, 3], 'b', label="Blue Line")

ax.set_title("Title")
ax.set_xlabel("X-Label")
ax.set_ylabel("Y-Label")
ax.legend()

plt.show()
```

## Colours
- `'r'` (single-letter shorthand): only 8 options
- `'red'` (named colors): 140 options - [Matplotlib Named Colors](https://matplotlib.org/stable/gallery/color/named_colors.html)
- `'#FF0000'` (Hex color codes): Use for precise branding or custom design - [Matplotlib Hex Color Codes](https://matplotlib.org/stable/gallery/color/named_colors.html)

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3, 4, 5], [2, 3, 1, 4, 5], color = 'r', label="Red")
ax.plot([1, 2, 3, 4, 5], [1,5,2, 4, 3], color='#2E8B57', label="Sea Green")

ax.set_title("Title")
ax.set_xlabel("X-Label")
ax.set_ylabel("Y-Label")
ax.legend()

plt.show()

# powertoys
```

## Line Styling

```python
# linewidth (lw) / transparency (alpha) / linestyle (ls)

fig, ax = plt.subplots()

ax.plot(x,y , color = 'purple', linewidth = 2, alpha = 1, linestyle = '--') # ls= '--', '-.' , ';'

plt.show()
```

```python
# marker / markersize / markerfacecolor / markeredgecolor / markeredgewidth

fig, ax = plt.subplots()

ax.plot(x,y , color = 'purple', linewidth = 3, alpha = 0.5, marker = '*', markersize = 20,
          markerfacecolor = 'yellow', markeredgecolor = 'blue', markeredgewidth = 1) # x, o , * , +)

plt.show()
```

## Adjust Axes
You can define the range for the axes:
- `ax.set_xlim([,])`
- `ax.set_ylim([,])`

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3, 4, 5], [1, 7, 2, 4, 3], 'b', label="Blue Line")

ax.set_xlim([2,5])
ax.set_ylim([0,12])

plt.show()
```

## Add Annotations
Adding text anywhere in the chart can be done via:
- `axes.text()` - this is using the same coordinates from the chart
- `fig.text()` - this is using the coordinates between 0-1

Both methods require at least three arguments:
- x: the position on the x-axis
- y: the position on the y-axis
- s: the text content





Adding Text via `ax.text()`

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3, 4, 5], [1, 7, 2, 4, 3])

ax.set_xlim([1,5])
ax.set_ylim([0,12])
ax.text(2,7.6, "Hello from the peak!")
ax.set_title("Title")
plt.show()
```

Adding Text via `ax.text()` and `fig.text()`

```python
x = [1,2,3,4,5]
y = [2,3,1,4,5]

fig, ax = plt.subplots()

ax.plot(x, y)
fig.text(0.15, 0.15, "Hi, I´m here at the top")
ax.text(2.5, 4, "Hi, I`m here in the plot.")

plt.show()
```
